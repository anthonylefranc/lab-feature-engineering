# Activity

Create a function that takes a dataframe as an input and a percentual threshold (default value = 0.25) and returns a list of columns with null values greater than the specified threshold

```python
def columns_with_high_nulls(df, threshold=0.25):
    """
input : dataframe & threshold
output : list of columns
    """
    # lenght of rows in the DataFrame
    total_rows = len(df)
    
    # number of null values to exceed the threshold
    min_nulls = total_rows * threshold
    
    # create boolean to indicate which columns have null values greater than the threshold
    columns_with_nulls = df.isnull().sum() > min_nulls
    
    # return column names above the threshold
    return columns_with_nulls[columns_with_nulls].index.tolist()
```

```python
high_nulls_list = columns_with_high_nulls(data)
print(high_nulls_list)
```
